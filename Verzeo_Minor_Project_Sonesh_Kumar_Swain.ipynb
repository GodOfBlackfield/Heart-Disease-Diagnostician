```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("heart.csv")
data
```

```python
target1 = data.iloc[0:254,13:14]
target1
```

```python
target2 = data.iloc[254:303,13:14]
target2
```

```python
train = data.iloc[0:254,0:13]
train
```

```python
test = data.iloc[254:303,0:13]
test
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
to_categorical(target1)
```

```python
#I had to create the target2 array manually as the target2 array was wrong with keras library (values were reversed with 
#respect to target1 array; also I shall name the target2 array as 'actual')
actual = np.zeros((49,2))
for x in range(0,49):
    actual[x] = np.array([1.,0.])
actual
```

# 1. Specify the architecture

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(70,activation = "relu",input_shape = (13,)))
model.add(Dense(2,activation = "Softmax"))
```

# 2. Compile the model

```python
model.compile(optimizer = "adam",loss = "sparse_categorical_crossentropy",metrics = ["accuracy"])
```

# 3. Fit the model

```python
model.fit(train,target1,epochs = 10000)
#Just for your information, I ran this command twice and I hope that it isn't against the rules of the project
```

# 4. Predict using the model

```python
prediction = model.predict(test)
prediction
```

```python
#This is to calculate the error found in prediction and actual target values
mse = []
for x in range(0,49):
    for y in range(0,2):
        if (actual[x][y] == 1):
            z = prediction[x][y]-actual[x][y]
            err = z*z
            mse.append(err)
error = np.mean(mse)
error
```
